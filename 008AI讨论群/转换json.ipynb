```python
import json
import os

def convert_dialogue_to_pretty_md(json_path):
    try:
        # 1. 动态生成文件名
        # os.path.basename 获取文件名（含后缀），os.path.splitext 将其拆分为 ('文件名', '.json')
        file_base_name = os.path.splitext(os.path.basename(json_path))[0]
        md_path = f"trans_data/tran_{file_base_name}.md"

        # 2. 读取 JSON 数据
        with open(json_path, 'r', encoding='utf-8') as f:
            data = json.load(f)

        messages = data if isinstance(data, list) else []
        
        # 3. 写入 Markdown
        with open(md_path, 'w', encoding='utf-8') as md_file:
            md_file.write(f"# 📝 对话档案记录\n\n")
            md_file.write(f"> **原始文件**：`{json_path}` | **总计轮次**：{len(messages)}\n\n")
            md_file.write("---\n\n")

            for msg in messages:
                speaker = msg.get("speaker", "未知发言者")
                content = msg.get("text", "").strip()
                timestamp = msg.get("timestamp", "")

                # 渲染排版
                md_file.write(f"#### 👤 **{speaker}**\n")
                if timestamp:
                    md_file.write(f"> 🕒 *{timestamp}*\n>\n")
                else:
                    md_file.write(f"> \n")

                formatted_content = content.replace("\n", "\n> ")
                md_file.write(f"> {formatted_content}\n\n")
                md_file.write("---\n\n")

        print(f"✅ 转换完成！已自动命名并生成：{md_path}")

    except Exception as e:
        print(f"❌ 发生错误：{e}")

if __name__ == "__main__":
    # 在这里输入你的 JSON 文件名
    filename = input("请输入 JSON 文件名：")
    target_json = "data/20260318-4/debate_history.json" 
    
    # 直接运行即可，函数内部会自动处理输出文件名
    convert_dialogue_to_pretty_md(target_json)
```
